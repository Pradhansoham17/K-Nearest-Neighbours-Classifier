```python
# after removing data row as index row and changing column names
import pandas as pd
import numpy as np
df=pd.read_csv('C:\\Users\\soham\\Downloads\\diabetes.csv',names=['Height','Weight','Diabetes'],header=None)
df  
```

```python
df['Height'].value_counts().idxmax()
```

```python
#performing train-test-split on our dataframe
# X consists of only categorical columns, y consists of output column("income")
import numpy as np
from sklearn.model_selection import train_test_split
X, y = df.iloc[:,0:2],df.iloc[:,[-1]]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3,shuffle=False)
```

```python
X_train
```

```python
X_test
```

```python
# calculate the Euclidean distance between two vectors
def euclidean_distance(row1, row2):
    distance = [(row1[i] - row2[i])**2 for i in range(len(row1))]
    return np.sqrt(np.sum(distance))

euclidean_distance(X_train.iloc[0], X_test.iloc[0])
```

```python
def getdistances(inp_traindf,test_row):
    distances=inp_traindf.apply(lambda row:euclidean_distance(row, test_row) ,axis=1)
    return distances
    
getdistances(X_train, X_test.iloc[0])
```

```python
def predictone(k,inp_traindf,test_row,train_opcol):
    ind=getdistances(inp_traindf,test_row).sort_values()[:k].index
    #print(train_opcol[ind])
    return train_opcol[ind].value_counts().idxmax()

predictone(3,X_train, X_test.iloc[0],y_train['Diabetes'])
```

```python
def predictall(k,inp_traindf,train_opcol,op_traindf):
    y_pred=op_traindf.apply(lambda row:predictone(k,inp_traindf,row,train_opcol), axis=1 )
    return y_pred

predictall(3,X_train,y_train['Diabetes'], X_test)   
```

```python
def accuracy(actlist,predlist):
    acc=(sum([a==b for (a,b) in zip(predlist,actlist)])/len(predlist))*100
    print('Accuracy of Algorithm is : ',acc,'%')
    return acc # used return bcoz we want to use it ahead in K-fold Cross Validation
    
accuracy(  y_test['Diabetes'].tolist()  ,  predictall(3,X_train,y_train['Diabetes'], X_test).tolist()  )
```

B) After applying feature scaling without using sklearn :

```python
dfs=df.copy()
dfs
```

```python
# Replacing the values in 'Height' and 'Weight' columns by their respective z-score values (Feature Scaling)
heightmean=dfs['Height'].mean()
weightmean=dfs['Weight'].mean()
heightstd=dfs['Height'].std()
weightstd=dfs['Weight'].std()

dfs['Height']=[(x-heightmean)/heightstd for x in dfs['Height']]
dfs['Weight']=[(x-weightmean)/weightstd for x in dfs['Weight']]

dfs
```

```python
#performing train-test-split on our dataframe
# X consists of only categorical columns, y consists of output column("income")
import numpy as np
from sklearn.model_selection import train_test_split
X1, y1 = dfs.iloc[:,0:2],dfs.iloc[:,[-1]]
X_train1, X_test1, y_train1, y_test1 = train_test_split(X1, y1, test_size=0.3,shuffle=False)
```

```python
X_train1
```

```python
X_test1
```

```python
accuracy(  y_test1['Diabetes'].tolist()  ,  predictall(3,X_train1,y_train1['Diabetes'], X_test1).tolist()  )
```
